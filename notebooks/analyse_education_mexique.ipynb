# Évolution des effectifs dans l'enseignement secondaire au Mexique


## Objectif
Analyser l’évolution du taux de scolarisation secondaire au Mexique.

## Données
Source : World Bank Open Data ( Indicateur : School enrollment, secondary (% gross) )

## Import des données

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/Projects/projet_education_Mexique/data/raw/School_enrollment_secondary_mx(world_bank).csv", skiprows=4, sep=",")
```

```python
df.head(10)
df.tail()

df.shape
#df.columns
```

## Filtrage des données pour le Mexique

```python
df_mexico = df[df["Country Name"] == "Mexico"]
df_mexico
```

```python
#df_mexico["1999"] NaN
#df_mexico["2000"] # valeurs à partir de 2000. La timeline commencera à partir de 2000 
```

```python
data_mexico = df_mexico.copy()
```

```python
data_mexico
```

## Nettoyage des données 

```python
serie = data_mexico.iloc[0] 
serie # une seule ligne de colonnes → transformée en liste verticale (mise en index)
```

```python
serie = serie.drop(["Country Name", "Country Code", "Indicator Name", "Indicator Code"])
```

```python
serie = serie.dropna()
serie
```

## Transformation des données 

```python
data_mexico = serie.reset_index() 
data_mexico
```

```python
data_mexico.columns = ["Année", "Valeur"]
```

```python
data_mexico.info()
```

```python
data_mexico["Année"] = pd.to_numeric(data_mexico["Année"])
data_mexico["Valeur"] = pd.to_numeric(data_mexico["Valeur"])
#print(data_mexico.dtypes)
```

## Export des données transformées 

```python
data_mexico.to_csv("../data/processed/mexico_processed.csv", index=False)
```

## Visualisation de l’évolution

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot(data_mexico["Année"], data_mexico["Valeur"], linewidth=2)

plt.title("Secondary School Enrollment in Mexico (%)")
plt.xlabel("Year")
plt.ylabel("Enrollment rate (%)")

plt.grid(True, alpha=0.3)

plt.axvline(2020, color="red", linestyle="--", alpha=0.7)
y_pos = data_mexico["Valeur"].max() * 0.9
plt.text(2020, y_pos, "COVID-19 pandemic", color="red")

plt.tight_layout()

plt.savefig("C:/Projects/projet_education_mexique/graphs/enrollment_trend.png")

plt.show()
```

## Analyse des résultats 

### Résultats principaux
- Au Mexique, le taux de scolarisation secondaire a augmenté sur la période étudiée, traduisant une amélioration de l’accès à l’éducation.  
- On observe cependant des fluctuations, notamment une baisse autour de 2020, pouvant être liée à la crise sanitaire mondiale du COVID-19 pandemic. 

### Interpretation 
Le dépassement du seuil de 100% (à partir 2015) s’explique par la nature de l’indicateur utilisé ("gross enrollment rate"), qui inclut des élèves en dehors de la tranche d’âge théorique (retard scolaire, redoublement).

Cela signifie que cet indicateur ne mesure pas directement la proportion d’enfants scolarisés, mais plutôt la pression globale sur le système éducatif.

### Limites

- Données manquantes sur certaines années
- Indicateur "gross" qui peut être trompeur
- Absence de distinction par genre ou zone géographique


# Conclusion

Malgré ses limites, l’analyse montre une amélioration globale de l’accès à l’éducation secondaire au Mexique, tout en soulignant la nécessité d’interpréter les indicateurs avec prudence.
